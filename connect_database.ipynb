```python
#create table
create_table_query = """
CREATE TABLE IF NOT EXISTS Telanagan_buses (
    id INT AUTO_INCREMENT PRIMARY KEY,
    Route_name VARCHAR(50),
    Route_link VARCHAR(100),
    Bus_name varchar(100),
    Bus_type varchar(100),
    departing_time time,
    Duration text,
    Reaching_time time,
    Star_rating float
);
"""
```

```python
cursor.execute(create_table_query)
```

```python
#Insert data into the table
data = pd.read_csv(r'C:/Users/Dell/OneDrive/Desktop/scraping/telanagana_buses.csv')
for index, row in data.iterrows():
    insert_query = """
    INSERT INTO Telanagan_buses (Route_name, Route_link, Bus_name, Bus_type, departing_time, Duration, Reaching_time, Star_rating)
    VALUES (%s, %s, %s, %s, %s, %s, %s , %s) 
    """
    cursor.execute(insert_query, (
        row['route_name'], row['route_link'], row['busname'], row['bustype'], 
        row['duparting_time'], row['duration'], row['reaching_time'],
        row['star_rating']
    ))
```

```python
result="SELECT id, Route_name, Route_link, Bus_name, Bus_type, departing_time, Duration, Reaching_time, Star_rating FROM Telanagan_buses;"
cursor.execute(result)
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python

df = pd.DataFrame(rows, columns=['id', 'Route_name', 'Route_link', 'Bus_name', 'Bus_type', 'departing_time', 'Duration', 'Reaching_time', 'Star_rating'])
df.to_csv(r'C:/Users/Dell/OneDrive/Desktop/streamlit/telanagana_buses1.csv', index=False)
```
