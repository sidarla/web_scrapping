```python
import pandas as pd
read_value=pd.read_csv("C:/Users/Dell/OneDrive/Desktop/scraping/telanagana_buses.csv")
read_value
```

```python
len(read_value.Price[0]),len(read_value.Price[1])
```

```python
for i in range(len(read_value.Price)):
  length_each_data=len(read_value.Price[i])
  if length_each_data>=7:
      read_value.Price[i]=read_value.Price[i][4:]
  elif length_each_data==3:
      read_value.Price[i]=read_value.Price[i]
```

```python
type(read_value.Price)
```

```python
read_value['Price'] = read_value['Price'].astype(int)
```

```python
read_value.info()
```

```python
# Convert to string, extract digits, and convert to integer
read_value['Seats_Available'] = read_value['Seats_Available'].astype(str)
read_value['Seats_Available'] = read_value['Seats_Available'].str.extract('(\d+)').astype(int)
```

```python
read_value.nunique()
```

```python
read_value.busname.unique()
```

```python
read_value.info()
```

```python
read_value['bustype']=read_value['bustype'].str.upper()
read_value
```

```python
read_value['bustype']= read_value['bustype'].apply(lambda x: 'NON A/C' if 'non a/c'.upper() in x else 'A/C')
read_value
```

```python
read_value.to_csv(r'C:/Users/Dell/OneDrive/Desktop/scraping/telanganabuses2.csv', index=False)
```
